```python
import json
from SolrClient import SolrClient
solr = SolrClient('http://localhost:8983/solr') 
import pandas as pd
```

```python
df = pd.read_csv('players.csv')
```

```python
docs = df.to_dict(orient='records')
```

```python
# indexing the JSON data in the Solr "FootballStatsCore" core
solr.index_json("FootballStatsCore", json.dumps(docs) )
solr.commit("FootballStatsCore", softCommit=True)
```

```python
# retrieve data from Solr
res = solr.query('FootballStatsCore', {
    'q': '*:*',
    'fl': '*',
    'rows': 100000,
})
result = res.docs

if(result == []):
    print("There are no players")
    print(len(result))
```
